```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cartopy

import matplotlib.patches as mpatches
import sandbox as sb
```

```python
from numpy import linspace, arange

def plot_edge(ax, lon0, lon1, lat0, lat1, n=200):
    """Draw a smooth edge between two points in lon/lat."""
    lons = linspace(lon0, lon1, n)
    lats = linspace(lat0, lat1, n)
    ax.plot(lons, lats,
            transform=ccrs.PlateCarree(),
            color=sb.utgray, linewidth=1.5)
```

```python
ax = plt.subplot(1,1,1,
    projection=ccrs.Orthographic(
        central_longitude=-20,   # center over Amundsen Sea
        central_latitude=35
    )
)
# figsize=(4,4), dpi=200
f=plt.gcf()
f.set_size_inches(8,8)
f.set_dpi(200)

# --- plot globe
ax.coastlines(linewidth=0.1)
ax.set_global()
ax.set_facecolor((93/256,114/256,167/256))
    # sb.utocean)

ax.add_feature(
    cfeature.LAND,
    facecolor=sb.utearth,
    zorder=1
)

# --- gridlines every 10 degrees
gl = ax.gridlines(
    crs=ccrs.PlateCarree(),
    draw_labels=False,
    linewidth=0.3,
    color='w',#sb.utgray,
    alpha=0.25,
    linestyle='-'
)

gl.xlocator = plt.FixedLocator(arange(-180, 181, 10))
gl.ylocator = plt.FixedLocator(arange(-90, 91, 10))

# --- make box
lon_min, lon_max = -10, -70
lat_min, lat_max = 15, 75

# --- 4 edges
plot_edge(ax, lon_min, lon_max, lat_min, lat_min)  # bottom
plot_edge(ax, lon_min, lon_max, lat_max, lat_max)  # top
plot_edge(ax, lon_min, lon_min, lat_min, lat_max)  # left
plot_edge(ax, lon_max, lon_max, lat_min, lat_max)  # right

plt.savefig('/Users/ivana/work/regionalgcm/img/baroc_ortho-projection-comparison.png')#, transparent=True)
plt.show()
```

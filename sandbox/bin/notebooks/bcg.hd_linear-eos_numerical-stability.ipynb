```python
from getSTDOUT import getData
from matplotlib import pyplot as plt
```

```python
# standard output files in $SCRATCH2/ivana/bcg.hd/run 
f120 = getData( 'out.fwd.120' )
f200 = getData( 'out.fwd.200' )
f240 = getData( 'out.fwd.240' )
```

```python
def plotBcgCFL(ax, cfl, title, ymax=0.06):
    sec = 'time_secondsf'
    
    ax.plot(f120.getVals( sec ), f120.getVals( cfl ), marker='.', label='dt = 120 s')
    ax.plot(f200.getVals( sec ), f200.getVals( cfl ), marker='.', label='dt = 200 s')
    ax.plot(f240.getVals( sec ), f240.getVals( cfl ), marker='.', label='dt = 240 s')
    ax.set_title(title)
    ax.set_xlabel('time [s]')
    if title == 'Zonal [UVEL]':
        ax.set_ylabel('CFL [1]')
    ax.set_ylim(0,ymax)
    ax.legend()
    ax.grid()
    
    return None
```

```python
# Plot CFL in u, v, and w for each unique time step size
u   = 'advcfl_uvel_max'; v   = 'advcfl_vvel_max'; w   = 'advcfl_wvel_max'

f, ax = plt.subplots(1,3, figsize=(12,3))
plt.suptitle('BCG HD: Advective Stability Monitoring', y=1.02)
plotBcgCFL(ax[0], u, 'Zonal [UVEL]')
plotBcgCFL(ax[1], v, 'Meridional [VVEL]')
plotBcgCFL(ax[2], w, 'Vertical [WVEL]')

plt.savefig('/home/ivana/regionalgcm/img/bcg.hd_cfl.png')
plt.show()
```

```python
uvel = 'dynstat_uvel_max'
vvel = 'dynstat_vvel_max'
wvel = 'dynstat_wvel_max'
sec  = 'time_secondsf'

f, ax = plt.subplots(1,3, figsize=(12,3))
plt.suptitle('BCG HD: Max Circulation', y=1.02)

ax[0].plot(f120.getVals( sec ), f120.getVals( uvel ), marker='.', label='dt = 120 s')
ax[0].plot(f200.getVals( sec ), f200.getVals( uvel ), marker='.', label='dt = 200 s')
ax[0].plot(f240.getVals( sec ), f240.getVals( uvel ), marker='.', label='dt = 240 s')
ax[0].set_title('Zonal [UVEL]')
ax[0].set_xlabel('time [s]')
ax[0].set_ylabel('speed [m/s]')
ax[0].legend()
ax[0].grid()

ax[1].plot(f120.getVals( sec ), f120.getVals( vvel ), marker='.', label='dt = 120 s')
ax[1].plot(f200.getVals( sec ), f200.getVals( vvel ), marker='.', label='dt = 200 s')
ax[1].plot(f240.getVals( sec ), f240.getVals( vvel ), marker='.', label='dt = 240 s')
ax[1].set_title('Meridional [VVEL]')
ax[1].set_xlabel('time [s]')
ax[1].legend()
ax[1].grid()

ax[2].plot(f120.getVals( sec ), f120.getVals( wvel ), marker='.', label='dt = 120 s')
ax[2].plot(f200.getVals( sec ), f200.getVals( wvel ), marker='.', label='dt = 200 s')
ax[2].plot(f240.getVals( sec ), f240.getVals( wvel ), marker='.', label='dt = 240 s')
ax[2].set_title('Vertical [WVEL]')
ax[2].set_xlabel('time [s]')
ax[2].set_ylim(0,0.0002)
ax[2].legend()
ax[2].grid()

plt.savefig('/home/ivana/regionalgcm/img/bcg.hd_vel.png')
plt.show()
```

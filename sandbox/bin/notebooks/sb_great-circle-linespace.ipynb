```python
import xarray as xr
import sandbox as sb

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tm4 = [-sb.degMinSec2decimal(70,54.0876), sb.degMinSec2decimal(39,50.8542)]
at  = [-sb.degMinSec2decimal(70,53.1670), sb.degMinSec2decimal(39,57.1500)]
```

```python
qq = sb.gcspace(tm4, at, npts=50)
```

```python
plt.figure()
plt.scatter(qq[0], qq[1])
plt.show()
```

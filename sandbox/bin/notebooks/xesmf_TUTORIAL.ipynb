```python
%matplotlib inline
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import numpy as np
import xarray as xr
import xesmf as xe
```

```python
ds = xr.tutorial.open_dataset('rasm') # use xr.tutorial.load_dataset() for xarray<v0.11.0
ds = ds.rename({'xc': 'lon', 'yc': 'lat'})
ds
```

```python
dr = ds['Tair']
dr
```

xe.util.grid_global(.5, .5) gives a 0.5$\circ \times$ 0.5$\circ$ resolution grid. See [xESMF](https://xesmf.readthedocs.io/en/latest/notebooks/Curvilinear_grid.html) for details.

```python
ds_out = xe.util.grid_global(.5, .5)
ds_out  # contains lat/lon values of cell centers and boundaries.
```

```python
regridder = xe.Regridder(ds, ds_out, 'bilinear')
```

```python
dr_out = regridder(dr)
```

```python
dr_out
```

```python
plt.figure(figsize=(12, 4))
ax = plt.axes(projection=ccrs.PlateCarree())
dr_out[0].plot.pcolormesh(ax=ax, x="lon", y="lat")
ax.coastlines()
```

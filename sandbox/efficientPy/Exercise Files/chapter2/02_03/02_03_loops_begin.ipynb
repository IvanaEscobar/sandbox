```python
cities, years = [], []

for game in open('games.txt','r'):
    words = game.split()
    
    city = ' '.join(words[:-1])
    year = words[-1].strip('()')
    
    cities.append(city)
    years.append(year)
```

Geopy can convert name locations to lat lon coordinates

```python
geolocator = geopy.geocoders.Nominatim(user_agent = 'test',\
                                       timeout=10)

locations = {}
for city in cities:
    print("Locating", city)
    locations[city] = geolocator.geocode(city.split('/')[0])
```

```python
locations['Paris']
```

```python
pp.figure(figsize=(10,5))

ax = pp.axes(projection=ccrs.PlateCarree())
ax.coastlines('110m')
ax.add_feature(cartopy.feature.BORDERS)
```

```python
pp.figure(figsize=(10,5))

ax = pp.axes(projection=ccrs.PlateCarree())
ax.coastlines('110m')
ax.add_feature(cartopy.feature.BORDERS)

for city,pos in locations.items():
    ax.plot(pos.longitude,pos.latitude,'r.',markersize=10)
```
